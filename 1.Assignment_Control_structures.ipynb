```python
bmi = float(input("Enter Body mass index: "))

if(bmi<18.5):
    print("Under Weight")
elif(bmi>=18.5 and bmi<=24.9):
    print("Healthy Weight")
elif(bmi>=25 and bmi<= 29.9):
    print("Over Weight")
elif(bmi>=30 and bmi<=34.9):
    print("Obese class 1")
elif(bmi>=35 and bmi<=39.9):
    print("Obese class 2")
else:
    print("Obese class 3")
```
