```python
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number

```

```python
import math

n = int(input("Number:"))
sqrN = int(math.sqrt(n))

flag = 0
for i in range(2,sqrN+1):
    if(n%i == 0):
        flag = 1

if(flag==0):
    print(n,"is a prime number")
else:
    print(n,"is not prime")
    
```
